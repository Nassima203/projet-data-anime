```python
# Charger le fichier CSV des anim√©s:

df = pd.read_csv('animes.csv')
print("Fichier charg√© avec succ√®s yeeeyy!")
print(f" Nombre d'anim√©s : {len(df)}")
```

```python
# Afficher les 5 premiers anim√©s
df.head()
```

```python
# Taille du tableau
lignes, colonnes = df.shape

print(f"Notre tableau contient :")
print(f"   ‚ûú {lignes} lignes ( {lignes} anim√©s)")
print(f"   ‚ûú {colonnes} colonnes (types d'informations)")
```

```python

#afficher les types de donn√©es et les valeurs manquantes par colonne

df.isnull().sum()

```

```python
#nettoyage des doublons
df.drop_duplicates(subset=['Anime'], keep='first')
print("Doublons supprim√©s avec succ√®s !")
print(f"il reste {len(df)} anim√©s dans le tableau apr√®s suppression des doublons.")
```

```python
# Mettre les noms d'anim√©s en majuscules
noms_majuscules = df['Anime'].str.upper()

print("Noms en majuscules :")
print(noms_majuscules.head(10).to_string())
```

```python
# Conversion de la colonne 'Date_Pub' en type datetime
df['Date_Pub'] = pd.to_datetime(df['Date_Pub'])

# Extraire l'ann√©e de publication
df['Annee'] = df['Date_Pub'].dt.year

print("üìÖ Dates converties et ann√©e extraite :")
print(df[['Anime', 'Date_Pub', 'Annee']].head(10))
```

```python
#Cr√©er une colonne "Is_Classic" si l'anim√© est sorti avant 2000
def verifier_classique(annee):
    if pd.isnull(annee):
        return "Inconnu"
    if annee < 2000:
        return "Oui"
    else:
        return "Moderne"
df['Is_Classic'] = df['Annee'].apply(verifier_classique)
# Afficher les anim√©s classiques
classiques = df[df['Is_Classic'] == "Oui"]
print(f"üéûÔ∏è {len(classiques)} anim√©s classiques (avant 2000) :")
print(classiques[['Anime', 'Annee', 'Is_Classic']].to_string(index=False))
```

```python
# Trier par Note_Globale (du plus haut au plus bas)
df_trie = df.sort_values('Note_Globale', ascending=False)

print("üèÜ Top 10 des anim√©s par note :")
print(df_trie[['Anime', 'Note_Globale', 'Studio']].head(10).to_string(index=False))
```

```python
# Filtrer les anim√©s avec une note sup√©rieure √† 8.5
excellents = df[df['Note_Globale'] > 8.5]

print(f"üåü {len(excellents)} anim√©s ont une note > 8.5 :")
print(excellents[['Anime', 'Note_Globale']].to_string(index=False))
```

```python
# Condition 1 : Sortie apr√®s 2020 (Modernit√©)
condition1 = df['Annee'] >= 2020

# Condition 2 : Format court (Binge-watchable en une soir√©e)
# la colonne correcte s'appelle 'Nb_Episodes'
condition2 = df['Nb_Episodes'] <= 13

# Application du filtre
p√©pites_r√©centes = df[condition1 & condition2]

# On trie par note pour avoir le meilleur contenu en haut
p√©pites_r√©centes = p√©pites_r√©centes.sort_values(by='Note_Globale', ascending=False)

print(f"üé¨ {len(p√©pites_r√©centes)} anim√©s r√©cents et courts trouv√©s :")
print(p√©pites_r√©centes[['Anime', 'Annee', 'Nb_Episodes', 'Note_Globale']].head(10).to_string(index=False))
```

```python
from pathlib import Path

print(f"‚úÖ Filtrage effectu√© : {len(df)} animes restants")
print(f"‚úÖ Medalist supprim√© : {'Medalist' not in df['Anime'].values}")

# S'assurer que le dossier de sortie existe
out_dir = Path('../data')
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
# === Cr√©ation Score de r√©gularit√© des notes d'√©pisodes ===

# Calcul de l'√©cart entre la meilleure et la pire note d'√©pisode
df['Ecart'] = df['Note_Meilleur_Ep'] - df['Note_Pire_Ep']
# Calcul du score de r√©gularit√©
df['Score_Regularite'] = 10 - df['Ecart']


print("\n" + "="*84 + "\n" + "TOP 5 - Animes les plus r√©guliers" + "\n" + "="*84)
print(df[['Anime', 'Note_Meilleur_Ep', 'Note_Pire_Ep', 'Ecart', 'Score_Regularite']].sort_values(by='Score_Regularite', ascending=False).head().to_string(index=False))
print("=" * 84 + "\n")
```

```python
# Note la plus haute et la plus basse
note_max = df['Note_Globale'].max()
note_min = df['Note_Globale'].min()

print(f" Note la plus haute : {note_max}")
print(f" Note la plus basse : {note_min}")

# Quel anim√© a la meilleure note ?
meilleur = df[df['Note_Globale'] == note_max]['Anime'].values[0]
print(f"\nüèÜ Le mieux not√© : {meilleur}")
```

```python
# Compter les anim√©s par statut
print("R√©partition par statut :")
print(df['Status'].value_counts())

print("\n" + "_"*40 + "\n")

# Top 5 des studios avec le plus d'anim√©s
print(" Top 5 des studios :")
print(df['Studio'].value_counts().head(5))
```

```python
# --- HISTOGRAMME COMPLET ET SIMPLE ---

# 1. On calcule la moyenne des notes
moyenne = df['Note_Globale'].mean()

# 2. On cherche les deux extr√™mes (le meilleur et le moins bon)
meilleur = df.loc[df['Note_Globale'].idxmax()]
pire = df.loc[df['Note_Globale'].idxmin()]

# 3. On pr√©pare la zone du graphique
plt.figure(figsize=(10, 6))

# 4. On dessine l'histogramme (les colonnes bleues)
# bins=20 : barres fines | rwidth=0.8 : espace entre les barres
plt.hist(df['Note_Globale'], bins=20, rwidth=0.8, color='yellow', edgecolor='white')

# 5. On ajoute la ligne de la MOYENNE (en rouge pointill√©)
plt.axvline(moyenne, color='red', linestyle='--', linewidth=2, label=f"Moyenne : {moyenne:.2f}")

# 6. On √©crit le nom du MEILLEUR et du MOINS BON


plt.text(meilleur['Note_Globale'], 2, f" {meilleur['Anime']}", color='green', fontweight='bold')
plt.text(pire['Note_Globale'], 2, f" {pire['Anime']}", color='red', fontweight='bold')

# 7. On ajoute les titres et la l√©gende
plt.title("O√π se situent vos anim√©s pr√©f√©r√©s ?", fontsize=14)
plt.xlabel("Note Globale")
plt.ylabel("Nombre d'anim√©s")
plt.legend() # Pour afficher l'√©tiquette de la moyenne

plt.show()
print(" La ligne rouge montre la moyenne")
```

```python
# === GRAPHIE DES STATUTS (FINI, EN COURS, EN PAUSE) ===

# 1. On compte combien d'anim√©s il y a pour chaque statut
statuts = df['Status'].value_counts()

# 2. On pr√©pare la zone du graphique
plt.figure(figsize=(8, 5))

# 3. On d√©finit de jolies couleurs (une pour chaque barre)

# rouge pour fini,  bleu pour en cours, gris pour en pause...
couleurs = ["#ff3700", '#6366f1', "#9C9E9E"]

# 4. On cr√©e les barres
bars = plt.bar(statuts.index, statuts.values, color=couleurs[:len(statuts)])

# 5. On ajoute le chiffre EXACT au-dessus de chaque barre
for bar in bars:
    hauteur = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, hauteur + 0.5, 
             int(hauteur), ha='center', fontweight='bold')

# 6. Titres et habillage
plt.title("Combien d'anim√©s par statut ?", fontsize=14, fontweight='bold')
plt.ylabel("Nombre d'anim√©s")
plt.grid(axis='y', linestyle='--', alpha=0.3) # Grille l√©g√®re derri√®re les barres

plt.show()
```

```python
# --- 1. PR√âPARATION ---
# On identifie les deux extr√™mes pour les mettre en valeur
meilleur = df.loc[df['Note_Globale'].idxmax()]
pire = df.loc[df['Note_Globale'].idxmin()]

# On cr√©e une version filtr√©e pour le second graphique (max 500 √©pisodes)
df_courts = df[df['Nb_Episodes'] <= 500]

# --- 2. GRAPHIQUE A : NOTE GLOBALE VS R√âGULARIT√â ---
plt.figure(figsize=(10, 6))

# On dessine tous les points en gris clair
plt.scatter(df['Note_Globale'], df['Score_Regularite'], c='lightgrey', alpha=0.5, label='Autres')

# On colore le top (Vert) et le bas (Rouge)
plt.scatter(meilleur['Note_Globale'], meilleur['Score_Regularite'], c='#22c55e', s=150, label='Top Note', edgecolors='black')
plt.scatter(pire['Note_Globale'], pire['Score_Regularite'], c='#ef4444', s=150, label='Note la plus basse', edgecolors='black')

# On ajoute les noms juste √† c√¥t√© des points
plt.text(meilleur['Note_Globale'], meilleur['Score_Regularite'], f"  {meilleur['Anime']}", weight='bold', va='center')
plt.text(pire['Note_Globale'], pire['Score_Regularite'], f"  {pire['Anime']}", weight='bold', va='center')

plt.title('1. Est-ce que les bons anim√©s sont r√©guliers ?', fontsize=14, fontweight='bold')
plt.xlabel('Note Globale')
plt.ylabel('Score de R√©gularit√©')
plt.legend()
plt.show()

# --- 3. GRAPHIQUE B : NOTE GLOBALE VS NOMBRE D'√âPISODES ---
plt.figure(figsize=(10, 6))

# Nuage de points violet pour voir si la longueur influence la note
plt.scatter(df_courts['Nb_Episodes'], df_courts['Note_Globale'], 
            c='#6366f1', alpha=0.6, s=80, edgecolor='white')

plt.title("2. Est-ce que les anim√©s longs sont mieux not√©s ?", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'√©pisodes (Limit√© √† 500)")
plt.ylabel("Note globale")
plt.grid(True, alpha=0.2) # Grille l√©g√®re pour mieux lire les chiffres
plt.show()

print(f"üí° Note moyenne g√©n√©rale : {df['Note_Globale'].mean():.2f}")
```

```python
#on ajoute une corr√©lation
# On regarde le lien entre la Note Globale et la Note du Meilleur √âpisode
corr = df['Note_Globale'].corr(df['Note_Meilleur_Ep'])

print(f"\nLien (Corr√©lation) entre Note Globale et Meilleur √âpisode : {corr:.2f}")

# Interpr√©tation simple du r√©sultat
if corr > 0.5:
    print(" INTERPR√âTATION : Corr√©lation POSITIVE forte !")
    print("   ‚Üí Un excellent √©pisode signifie souvent que l'anim√© entier est excellent.")
elif corr > 0:
    print(" INTERPR√âTATION : Corr√©lation POSITIVE faible.")
    print("   ‚Üí Il y a un lien, mais un bon √©pisode ne garantit pas tout l'anim√©.")
else:
    print("INTERPR√âTATION : Corr√©lation N√âGATIVE.")
    print("   ‚Üí √âtonnant : les bons √©pisodes ne semblent pas aider la note globale.")

# --- 4. R√âSUM√â DU CATALOGUE ---
print(f"\nüí° Note moyenne des anim√©s : {df['Note_Globale'].mean():.2f}")
print(f"üí° Score de r√©gularit√© moyen : {df['Score_Regularite'].mean():.2f}")
```
